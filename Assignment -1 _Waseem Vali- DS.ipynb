In this part, you will be working with the library ```pandas```. Import this library with an alias name ```pd```.

````
# Initial imports
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv
Read the data and display the first 5 rows

````
#Download dataset from: http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv

file_url = "http://biostat.mc.vanderbilt.edu/wiki/pub/Main/DataSets/titanic3.csv"
df = pd.read_csv(file_url)
df.head()
````

### Preprocessing of the Dataset



---


**1. Convert the categorical values of columns 'sex', 'embarked', 'pclass' into dummy values and add them into titanic_data_new DataFrame**




---



````
titanic_data_newdf = pd.get_dummies(df[['sex', 'embarked', 'pclass']])
titanic_data_newdf.head()
````



---


**2. Fill the missing data in Age and Fare columns with the mean value of the respected column then add these columns to titanic_data_new**


---



````
mean_age = df['age'].mean()
mean_fare = df['fare'].mean()
titanic_data_newdf['age'] = df['age'].fillna(mean_age)      #fill empty spaces with mean values and add it to new df
titanic_data_newdf['fare'] = df['fare'].fillna(mean_fare)       #fill emplty spaces with mean values and add it to new df
titanic_data_newdf.head()
````

###Normalization


---



1. Normalize the columns which have numerical data



---



````
for column in df.columns:                #iterating over columns with numerical data types
  if df[column].dtypes == 'int64' or df[column].dtypes =='float64':
    df[column]= df[column]/(df[column].max())               # normalizing by dividing the column by mean of the column
df.head()
````



---
2. Drop the original age, fare columns
3. Append the normalized columns to the titanic_data_new DataFrame.



---



````
df= df.drop(columns = ['age', 'fare'])

for column in df.columns:                                              #appending columns with numerical data which was normalized in previous steps to titanic_data_newdf dataframe
  if df[column].dtypes == 'int64' or df[column].dtypes =='float64':
    titanic_data_newdf[column]= df[column]

titanic_data_newdf.head()
````

###Plotting
1. plot the age against the ID(continuous)
2. plot the fare against the ID(scatter)

````
titanic_data_newdf['ID']= titanic_data_newdf.index                 #setting index a column in the dataframe
titanic_data_newdf.plot(x='ID', y='age', figsize = (8,4), fontsize = 14)
titanic_data_newdf.plot(x='ID', y='fare', kind='scatter', figsize = (8,4), fontsize = 14)
````

Plot the sine function with domain in $[-3\pi,3\pi]$. Label the axes appropriately. Add title $\sin(x)$. Use latex formatting while adding the title. Set the plot size to (10,10)

````
%matplotlib inline
from math import pi
x = np.arange(-3*pi, 3*pi, 0.01)
y = np.sin(x)
plt.rcParams.update({'font.size': 18})
plt.figure(figsize=(15,5))
plt.plot(x,y)
plt.title(r'$\sin(x)$ Function')
plt.xlabel(r'$x$')
plt.ylabel(r'$\sin(x)$')
plt.grid(True, which='both')

````

Plot $\sin(x)$ and its spectrum, $\cos(x)$ and its spectrum in a single plot, in the domain $[-2\pi,2\pi]$. Label the plot appropriately.

````
x = np.linspace(-2*np.pi, 2*np.pi, 256)
y1 = np.sin(x)                        #sine function
y2 = np.cos(x)                        #cosine function
y3 = np.fft.fft(y1)                   #fourier transform of sine function
y4 = np.fft.fft(y2)                   #fourier transfomr of cosine function
freq = np.fft.fftfreq(x.size)         #calculate freq domain for fourier tranform 
figure, ax = plt.subplots(4,1, figsize=(8,12))
plt.subplots_adjust(hspace=0.5)

#plotting sine 
ax[0].set_title('Sine wave')
ax[0].plot(x, y1)
ax[0].set_xlabel(r'$x$')
ax[0].set_ylabel(r'$\sin(x)$')

#fourier transform of sine
ax[1].set_title('Fourier Transform of Sine')
ax[1].plot(freq, y3.real)
ax[1].set_xlabel(r'$freq$')
ax[1].set_ylabel(r'Fourier Transform of $\sin(x)$')

#plotting cosine
ax[2].set_title('Cosine wave')
ax[2].plot(x, y2)
ax[2].set_xlabel(r'$x$')
ax[2].set_ylabel(r'$\cos(x)$')

#fourier transform of cosine
ax[3].set_title('Fourier Transform of Cosine')
ax[3].plot(freq, y4.real)
ax[3].set_xlabel(r'$freq$')
ax[3].set_ylabel(r'Fourier Transform of $\cos(x)$')
````
